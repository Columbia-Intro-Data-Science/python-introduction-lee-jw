```python
import requests
import pandas as pd
import requests
import numpy as np
import time
```

```python
url = "https://api.icowatchlist.com/public/v1/finished"
data = requests.get(url).json()["ico"]["finished"]
df = pd.DataFrame(columns = list(['name','start_time', 'price_usd', 'all_time_roi']))
for i in range(len(data)-1):
    df = df.append({'name': data[i]['name'], 'price_usd': data[i]['price_usd'], 'start_time': data[i]['start_time'], 'all-time_roi': data[i]['all_time_roi']}, ignore_index=True)
```

```python
print(df)
```

```python
df.to_csv('additional_data.csv')
```
